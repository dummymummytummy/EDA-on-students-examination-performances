```python

#importing numpy pandas mathplotand other methods
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#first read the data file
df= pd.read_csv('StudentsPerformance.csv')
```

```python
# quick look to dataset
df.head(30)
```

```python
#Print summary statistics
df.describe()
```

```python
#the complete information about the dataset
df.info()
```

```python


# Inspect missing values in the dataset
print(df.isnull().values.sum())
```

```python
# seaborn histogram 
sns.distplot(df['math score'], hist=True, kde=False, 
             bins=12, color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Score count of maths subject')
plt.xlabel('maths')

plt.ylabel('score')
```

```python
# seaborn Scatterplot
sns.scatterplot(x=df['race/ethnicity'], y=df['math score'])
```

```python
# line plot 
sns.lineplot(x='race/ethnicity',y='math score', data=df )
```

```python
#box plot
sns.boxplot(y = 'gender',data= df, x= 'reading score')
plt.xlabel('reading score')
```

```python
# seaborn Scatterplot
sns.scatterplot(x=df['race/ethnicity'], y=df['reading score'])
```

```python
# seaborn histogram 
sns.distplot(df['reading score'], hist=True, kde=False, 
             bins=12, color = 'green',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Reading Score count')
plt.xlabel('reading')

plt.ylabel('score')
```
